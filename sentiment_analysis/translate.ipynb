```python
# %pip install sacremoses
import pandas as pd
from transformers import MarianMTModel, MarianTokenizer

# num = 4

model_name = "Helsinki-NLP/opus-mt-nl-en"
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)

# df = pd.read_pickle(f'./persistent_translator.df{num}.pkl')
df = pd.read_pickle('./hitteplan_persistent_translator.df.pkl')

length = len(df)

for index, row in df.iterrows():
    
    text = row['comment_text']
    
    if row['translated_comment'] == None:
        
        inputs = tokenizer(text, return_tensors="pt", truncation=True)
        translation = model.generate(**inputs)
        translated_text = tokenizer.batch_decode(translation, skip_special_tokens=True)[0]
        df.loc[index, 'translated_comment'] = translated_text
        
        print(f'{index+1}/{length} complete | {(index + 1) / length * 100}% done')
        
        df.to_pickle('./hitteplan_persistent_translator.df.pkl')

print("Done!")
```
